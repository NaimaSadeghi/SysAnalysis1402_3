## Histogram

```python
# Rice rule for finding the number of bins
n=10000
b=2*n**(1/3)
b
```

```python
import matplotlib.pyplot as plt
import random
n=10000
b=2*n**(1/3)
def r():
    return random.random()*6
val=[]
for i in range(10000):
    val.append(r()+r())
plt.hist(val,bins=40,ec='k')
plt.xlabel("Outcome")
plt.ylabel("Number of Occurrences")
```
