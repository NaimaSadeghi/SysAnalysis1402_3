```python
import numpy as np
def project_duration(a,b,c,d,e):
    return a+np.maximum(b,c)+d+e
```

```python
import simpm.dist as dist
a=dist.triang(2,3,6).samples(100000)
b=dist.triang(2,3,6).samples(100000)
c=dist.triang(2,3,6).samples(100000)
d=dist.triang(2,3,6).samples(100000)
e=dist.triang(3,5,6).samples(100000)

vals=project_duration(a,b,c,d,e)
```

```python
import matplotlib.pyplot as plt
_=plt.hist(vals,bins=40)
```

```python
np.percentile(vals,90)
np.std(vals)
```

## Computing $\pi$

```python
number_of_runs=1000000
x=dist.uniform(-1,1).samples(number_of_runs)
y=dist.uniform(-1,1).samples(number_of_runs)
distance=x**2+y**2
```

## Simulate waiting time for a bus


```python
import random
import simpm.dist as dist
import matplotlib.pyplot as plt
import numpy as np
val=[]

for i in range(10000):
    Bus_arrivaltime=dist.expon(5).sample()
    Our_arrivaltime=random.random()*15
    waiting_time=Bus_arrivaltime-Our_arrivaltime
    while waiting_time<0:
        Bus_arrivaltime+=dist.expon(5).sample()
        waiting_time=Bus_arrivaltime-Our_arrivaltime
    val.append(waiting_time)
```

```python


fig,ax=plt.subplots(1,1,figsize=(5,5))
_=plt.hist(val,bins=40,density=True)
ax.set_xlabel("waiting time")
val=np.array(val)
np.mean(val)
```
