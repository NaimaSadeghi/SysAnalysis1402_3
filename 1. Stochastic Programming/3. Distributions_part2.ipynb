# Distributions (part 2)

## Normal
<img src="./Figures/normal_pdf.png" width="400px" /> 
<img src="./Figures/normal_sym.png" width="400px" /> 

## Alternative names
* Bell shaped curve
* Gaussian distribution

```python
# generate samples from a normal distribution
import random
import matplotlib.pyplot as plt
mu=4
sigma=3
val=[]
for i in range(1000000):
    s=random.gauss(mu,sigma)
    val.append(s)
count,bins,_=plt.hist(val,bins=50,density=True)
```

```python
import numpy as np
mu=4
std=7
val=np.random.normal(mu,std,10000000)
count,bins,_=plt.hist(val,bins=100,density=True)
plt.plot((bins[1:]+bins[:-1])/2,count,color="red")
plt.ylabel("f(x)")
plt.xlabel("x")
plt.title("Normal PDF")
```

```python
count2,bins2,_=plt.hist(val,bins=100,density=True,cumulative=True)
plt.plot((bins2[1:]+bins2[:-1])/2,count2,color="red")
plt.ylabel("f(x)")
plt.xlabel("x")
plt.title("Normal CDF")
```

## Confidence interval

<img src="./Figures/normal_confidence.png" width="400px" /> 

```python
# find confidence interval for 80% confidence
int_max=np.percentile(val,90)
int_min=np.percentile(val,10)
print(f"the interval is: {int_min:.1f},{int_max:.1f}")
```

## Uniform Distribution

```python
import numpy as np
a=4
b=7
val=np.random.uniform(a,b,10000000)
count,bins,_=plt.hist(val,bins=100,density=True)
plt.plot((bins[1:]+bins[:-1])/2,count,color="red")
plt.ylabel("f(x)")
plt.xlabel("x")
plt.title("Uniform PDF")
plt.ylim(0,.4)
```

```python
count2,bins2,_=plt.hist(val,bins=100,density=True,cumulative=True)
plt.plot((bins2[1:]+bins2[:-1])/2,count2,color="red")
plt.ylabel("f(x)")
plt.xlabel("x")
plt.title("Uniform CDF")
```

# Triangular


```python
import numpy as np
low=4
maximum=7
mode=5
val=np.random.triangular(low,mode,maximum,10000000)
count,bins,_=plt.hist(val,bins=100,density=True)
plt.plot((bins[1:]+bins[:-1])/2,count,color="red")
plt.ylabel("f(x)")
plt.xlabel("x")
plt.title("Triangular PDF")
```

```python
count2,bins2,_=plt.hist(val,bins=100,density=True,cumulative=True)
plt.plot((bins2[1:]+bins2[:-1])/2,count2,color="red")
plt.ylabel("F(x)")
plt.xlabel("x")
plt.title("Triangular CDF")
```

## Exponential Distribution 

```python
import numpy as np
mu=20
val=np.random.exponential(mu,10000000)
count,bins,_=plt.hist(val,bins=100,density=True)
plt.plot((bins[1:]+bins[:-1])/2,count,color="red")
plt.ylabel("f(x)")
plt.xlabel("x")
plt.title("Exponential PDF")
```
