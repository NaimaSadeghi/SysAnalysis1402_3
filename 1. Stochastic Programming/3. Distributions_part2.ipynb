# Distributions

### Class Exe.

* Bus arrives every (0,15) minutes 
* We go to bus stop randomly
* Write a program that plots the histogram of our waiting time




```python
import random
import matplotlib.pyplot as plt
val=[]

for i in range(10000):
    Bus_arrivaltime=random.random()*30
    Our_arrivaltime=random.random()*15
    waiting_time=Bus_arrivaltime-Our_arrivaltime
    while waiting_time<0:
        Bus_arrivaltime+=random.random()*15
        waiting_time=Bus_arrivaltime-Our_arrivaltime
    val.append(waiting_time)
plt.hist(val,bins=40,ec="k")

```
