```python
import pandas as pd
```

```python

mydf=pd.read_csv("force_length.csv")

mass=mydf["mass"].values
length=mydf["length"].values
```

```python
mydf.plot.scatter(x="mass",y="length")
```

```python
import numpy as np
#ax+b
#a*mass+b=length
a,b=np.polyfit(mass,length,1)
```

```python
predicted_length=a*mass+b
```

```python
len(length)
```

```python
#mean square error

mse=np.sum((predicted_length-length)**2)/len(predicted_length)

rmse=np.sqrt(mse)
rmse
```

```python
a,b,c=np.polyfit(mass,length,2)
predicted_length=a*mass**2+b*mass+c
mse=np.sum((predicted_length-length)**2)/len(predicted_length)
rmse=np.sqrt(mse)
rmse
```

```python
a,b,c,d=np.polyfit(mass,length,3)
predicted_length=a*mass**3+b*mass**2+c*mass+d
mse=np.sum((predicted_length-length)**2)/len(predicted_length)
rmse=np.sqrt(mse)
rmse
```
